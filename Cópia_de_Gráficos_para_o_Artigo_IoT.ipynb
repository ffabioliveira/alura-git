```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
### Criar figura e eixos ###
fig, ax = plt.subplots(figsize = (20, 10))
fig_aplicacao = 'distribuicao_artigos_revisados_APLICACAO.png'
xlab = 'Aplicações'
ylab = 'Quantidade'
fsize, lsize = 22, 22
#Distribuição dos artigos revisados ​​de acordo com os domínios de aplicação

#######################
distribuicao_aplicacao = np.array(['Agricultura\n de precisão',
'Ciência de dados,\nBig data e \naprendizagem\nde máquina',
'Movido a\nenergia solar',
'Controle de \nqualidade do\ncrescimento \nde safras',
'Controle \nde doenças'])

### eixo y aplicações distribuidas ###
qtd = np.array([13, 2, 2, 1, 1])

### Plotar dados ###
ax.bar(distribuicao_aplicacao, qtd, color = 'red')
### Configurar gráfico ###
ax.set_xlabel(xlab, fontsize = fsize)
ax.set_ylabel(ylab, fontsize = fsize)
ax.tick_params(labelsize = lsize, width = 2)

for i in range(5):
	plt.text(distribuicao_aplicacao[i], qtd[i], qtd[i], ha = 'center', va = 'bottom', fontsize = fsize)

### Salvar a Figura ###
fig.savefig(fig_aplicacao)
plt.show()
```
